## To replace '?' with means values in simplest way i will use SimpleImputer from sklearn
#### First we convert '?' to np.nan because simple imputer doesn't recognize it as object 
#### bydeafut np.nan is float so we replace '?' with np.nan first then we pass it in simple imputer

```python
df['Protein (g)'].replace('?',np.nan,inplace=True)
df['Sugar (g)'].replace('?',np.nan,inplace=True)
df['Trans Fat (g)'].replace('?',np.nan,inplace=True)
```

```python
# for checking its got replace or not
df.info()
print(("*")*125)
print("\t\t\t\t**************ANALYSIS***************\n\n As we see in above info the dtypes of few columns is objects but now they will get replced with their mean values and now their dtypes is float")
print("\n\n THE MISSED DATA TYPES COLUMNS ARE : \n\n 6   Protein (g)        530 non-null    float64 \n 8   Sugar (g)          530 non-null    float64 \n 11  Trans Fat (g)      530 non-null    float64")
print(("*")*125)
```

## 1 - Predicting the data using Linear Regression

```python
# Importing the LinearRegression library from sklearn
from sklearn.linear_model import LinearRegression

# assigning a variable as lr
lr=LinearRegression()

#fitting the lr with train and test  data
lr.fit(xtrain,ytrain)

# for prdiction of new values of our targeted columns we use below codes
ypred=lr.predict(xtest)
```

```python
# Predecting the new values of our targeted columns 
ypred
```

### For getting the values as ytest and ypred errors
### For mininizing the errors in our model traing we will use
#### => mean_absolute_error
#### => mean_squared_error
#### => r2_score

```python
# For checking the ERRORS in our model
print(ytest-ypred)
```

```python
# Importing the libraries to minimize the errors in our ytest and ypred 
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

# Assigning a mae as variable for mean_absolute_error
mae=mean_absolute_error(ytest,ypred)

# Assigning a mse as variable for mean_squared_error
mse=mean_squared_error(ytest,ypred)

# Assigning a rmse as variable for sqrt(mse)
rmse=np.sqrt(mse)

# Assigning a r2 as variable for r2_score
r2=r2_score(ytest,ypred) # this will gives us the accuracy of our model

# Printing the value of all above tests
print(f'mae : {mae}\nmse : {mse}\nrmse : {rmse}\naccuracy : {r2}')
```

## We have achieved an Accuracy of 97 % which is good Enough.

# 2 - Predicting the data using Decision Tree
### Decision Tree Regressor

```python
# Importing the library from sklearn for decission tree
from sklearn import tree

# after importing tree we import now our DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
# Passing our model train and test data in function 
def mymodel(model):
   model.fit(xtrain,ytrain)
   ypred=model.predict(xtest)
   print(r2_score(ytest,ypred))
```

```python
#Assigning a dt as variable for DecisionTreeRegressor
dt=DecisionTreeRegressor()
```

```python
# decission tree will gives each and every time a different accuracy as we run it because it will select random data in operation
mymodel(dt)
```

## By using Decision Tree we get an accuracy of 92% which is good but lets check whether we get more accurcy by hyper tunning ploting decision tree

# PLOTTING A DECISSION TREE DIAGRAM

```python
# for ploting a decission tree we use chart 
plt.figure(figsize=(25,10))
chart=tree.plot_tree(dt,max_depth=5,feature_names=x.columns,filled=True,fontsize=10)
plt.savefig('abc.jpg')
```

# Train accuracy is high and Test acc is low
### (low bias high varience)

# HYPER PARAMETER TUNNING (HPT)
### We use HPT to overcome this problem
### i) - max_depth

```python
# passing train and test data in function to find out our best max depth values
for i in range(1,50):
    dt1= DecisionTreeRegressor(max_depth=i)
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)
    
    rs=r2_score(ytest,ypred)
    # for printing the scores 
    print(f"max_depth {i}  r2_score {rs}")
```

```python
# for new accuracy from model by max_depth
dt2=DecisionTreeRegressor(max_depth=3)
mymodel(dt2)
```

#### For Train data accuracy

```python
# Accuracy after passing xtrain and ytrain data
dt2.score(xtrain,ytrain)
```

#### For Test data accuracy

```python
# Accuracy after passing xtest and ytest data
dt2.score(xtest,ytest)
```

## By using Decision Tree Max Depth function we achive the accuracy of 78%
### ii) - min_samples_split

```python
# by using foreloop passing train and test data for accuracy of model
for i in range(2,50):
    dt3= DecisionTreeRegressor(min_samples_split=i)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    
    rs=r2_score(ytest,ypred)
    
    print(f"min_samples_split {i}  r2_score {rs}")
```

```python
# for new accuracy from model by min_samples_split methods
dt4=DecisionTreeRegressor(min_samples_split=4)
mymodel(dt4)
```

```python
# Accuracy after passing xtrain and ytrain data
dt4.score(xtrain,ytrain)
```

```python
# Accuracy after passing xtest and ytest data
dt4.score(xtest,ytest)
```

## By using Decision Tree Min Sample Split function we achive the accuracy of 84%
###  iii) - min_samples_leaf

```python
# by using foreloop passing train and test data for accuracy of model
for i in range(1,50):
    dt5= DecisionTreeRegressor(min_samples_leaf=i)
    dt5.fit(xtrain,ytrain)
    ypred=dt5.predict(xtest)
    
    rs=r2_score(ytest,ypred)
    
    print(f"min_samples_split {i}  r2_score {rs}")
```

```python
# for new accuracy from model by min_samples_leaf methods
dt6=DecisionTreeRegressor(min_samples_leaf=5)
mymodel(dt6)
```

```python
# Accuracy after passing xtrain and ytrain data
dt6.score(xtrain,ytrain)
```

```python
# Accuracy after passing xtest and ytest data
dt6.score(xtest,ytest)
```

## By using Decision Tree min_samples_leaf function we achive the accuracy of 89%

# 3 - Predecting the data using Random Forest

```python
# Importing the RandomForestRegressor library from sklearn
from sklearn.ensemble import RandomForestRegressor

## Importing the train_test_split library from sklearn.model_selection
from sklearn.model_selection import train_test_split
rfr = RandomForestRegressor(n_estimators=100, max_depth=5)

rfr.fit(xtrain, ytrain)
ypred = rfr.predict(xtest)
```

```python
# Importing the libraries to minimize the errors in our ytest and ypred 
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print('MSE:', mean_squared_error(ytest, ypred))
print('MAE:', mean_absolute_error(ytest, ypred))
print('r2_score:', r2_score(ytest, ypred))
```
